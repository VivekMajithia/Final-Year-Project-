```python
import tensorflow_datasets as tfds
import tensorflow as tf
from skimage import filters
import matplotlib.pyplot as plt
dataset, info = tfds.load('cats_vs_dogs', split='train', with_info=True, as_supervised=True)
sample_image, _ = next(iter(dataset))
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.imshow(sample_image.numpy())
plt.title('Original Image')
denoised_image = filters.median(sample_image.numpy().astype('uint8'))
plt.subplot(1, 2, 2)
plt.imshow(denoised_image)
plt.title('Denoised Image')

plt.show()
```
