```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt

dataset, info = tfds.load('cats_vs_dogs', with_info=True, as_supervised=True)

train_ds = dataset['train']
for image, label in train_ds.take(22):
    second_image_train = image

test_ds = train_ds.take(2)

plt.imshow(second_image_train)
plt.title(f"Training Label: {label.numpy()}")
plt.show()

for image, label in test_ds:
    second_image_test = image
    break

plt.imshow(second_image_test)
plt.title(f"Testing Label: {label.numpy()}")
plt.show()


#Here we are just getting the images and seeing the labels to ensure we get dogs/cats, this is just a test
```
