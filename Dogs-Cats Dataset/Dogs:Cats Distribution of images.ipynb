```python
import tensorflow as tf
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
dataset, info = tfds.load('cats_vs_dogs', split='train', with_info=True, as_supervised=True)
num_cat_examples = info.splits['train'].num_examples // 2
num_dog_examples = num_cat_examples
plt.bar(['Cat', 'Dog'], [num_cat_examples, num_dog_examples])
plt.xlabel('Class')
plt.ylabel('Number of Examples')
plt.title('Distribution of Cat and Dog Images in the Dataset')
plt.show()
cat_samples = dataset.take(num_cat_examples)
dog_samples = dataset.skip(num_cat_examples).take(num_dog_examples)

plt.figure(figsize=(10, 5))
for i, (cat_img, dog_img) in enumerate(zip(cat_samples, dog_samples), 1):
    plt.subplot(2, num_cat_examples, i)
    plt.imshow(cat_img[0])
    plt.title('Cat')

    plt.subplot(2, num_dog_examples, i + num_cat_examples)
    plt.imshow(dog_img[0])
    plt.title('Dog')

plt.show()
```
