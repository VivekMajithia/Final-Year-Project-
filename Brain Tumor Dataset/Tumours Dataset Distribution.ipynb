```python
import zipfile
import os
import matplotlib.pyplot as plt
zip_file_path = '/content/drive/MyDrive/Colab Notebooks/archive (1).zip'
extracted_dir = '/content/drive/MyDrive/Colab Notebooks'
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir)
dataset_path = os.path.join(extracted_dir, 'Data')
class_counts = {}
for class_name in os.listdir(dataset_path):
    class_path = os.path.join(dataset_path, class_name)
    if os.path.isdir(class_path):
        num_images = len(os.listdir(class_path))
        class_counts[class_name] = num_images
sorted_classes = sorted(class_counts.items(), key=lambda x: x[1], reverse=True)
class_names = [x[0] for x in sorted_classes]
class_counts = [x[1] for x in sorted_classes]
plt.figure(figsize=(10, 6))
plt.barh(class_names, class_counts, color='skyblue')
plt.xlabel('Number of Images')
plt.ylabel('Class')
plt.title('Distribution of Classes in Dataset')
plt.gca().invert_yaxis()
plt.show()
```
